# Relationship between Beds/Baths to Price

```python
plt.figure(figsize=(8, 6))
sns.relplot(data=df[(df['beds'] < 9) & (df['price'] < 100000)], y='price', x='beds')
plt.show()
```

```python
def get_average_rent(state: str, data: pd.DataFrame) -> float:
    """
    Calculates the average rent price for a specific state.

    Parameters:
    - state: The state for which to calculate the average rent.
    - data: The dataset (DataFrame) containing the housing information.

    Returns:
    - The average rent price for the specified state.
    """
    state_data = data[data['state'] == state]
    average_rent = state_data['price'].mean()

    return average_rent
```

```python
state = 'fl'  # You can change this to any state code you want
average_rent = get_average_rent(state, df)

print(f"The average rent for {state.upper()} is: ${average_rent:,.2f}")
```
